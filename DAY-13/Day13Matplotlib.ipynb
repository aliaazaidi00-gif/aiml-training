```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(10)
y=np.random.rand(10)
temp=np.random.randint(20,30,size=10)
plt.scatter(x,y,c=temp,cmap='plasma', s=120, edgecolor='black')
plt.colorbar(label='Temperature')
plt.title('Scattered plot Colored by Temperature')
plt.xlabel('X-Random')
plt.ylabel('Y-Random')
plt.show()
```

```python
height_cm=[150,155,160,165,170,175,180,185]
weight_kg=[50,53,56,59,62,66,68,72]
age_year=[20,22,25,28,30,35,38,40]
```

```python
plt.figure(figsize=(8,5))
```

```python
scatter=plt.scatter(height_cm, weight_kg,c=age_year,cmap='rainbow',s=120, alpha=0.9)
plt.colorbar(scatter,label='Age (years)')
plt.title('Scattered plot: Height vs Weight (colored by age)')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(color='blue', linestyle='--', linewidth=0.7)
plt.show()
```

# Pie Plot Example
    ## plt.pie(x,labels='', colors='', autopct='', startangle='', shadow=, explode=, radius) 

```python
depts=['HR','Web-Development','App-Development','Agile-Scrum','Testing']
expenses=[30,10,30,15,15]
```

```python
plt.figure(figsize=(7,7))
```

```python
plt.pie(expenses, labels=depts, autopct='%1.0f%%', startangle=140, explode=(0, 0.05, 0, 0, 0),shadow=True)
plt.title('Department wise Expense Report')
plt.show()
```

```python
plt.pie(expenses, labels=depts, colors=plt.cm.pink([0.1,0.3,0.5,0.7,0.9]), autopct='%1.0f%%', startangle=140, explode=(0, 0.05, 0, 0, 0),shadow=True)
plt.title('Department wise Expenses Report')
plt.show()
```

# Create a pie plot using following data
    sales_data={Mobile':45000, 'Clothing':30000, 'Groceries':35000, 'Home Decor':20000}
    show percentage with two decimal
    use colormap to give the difference color to your pie

```python
sales_data={'Mobile':45000, 'Clothing':30000, 'Groceries':35000, 'Home Decor':20000}
cat=list(sales_data.keys())
sale=list(sales_data.values())
print(cat)
print(sale)
plt.figure(figsize=(7,7))
plt.pie(sale, labels=cat, colors=plt.cm.coolwarm([0.6,0.3,0.5,0.7]), textprops={'color':'black','fontsize':10, 'fontweight':'bold'},
        autopct='%1.2f%%', startangle=140, explode=(0, 0.05, 0, 0),shadow=True, radius=1.2)
plt.title('Categories Wise Sale Report')
plt.legend(loc='lower left')
plt.show()
```

# Histogram Example

```python
marks=[45,60,80,55,62,90,89,50,56,60,80,55,62,90,89,58,90,56]
plt.style.use('ggplot')
plt.figure(figsize=(8,5))
plt.hist(marks,bins=5,color='maroon', edgecolor='black', alpha=0.8)
plt.title('Histogram for Student Marks')
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')
plt.grid(color='grey', linestyle='--', linewidth=0.7)
plt.show()
```
