```python
import matplotlib.pyplot as plt
```

```python
months=['Jan','Feb','March','April']
sales=[2000, 3000, 2500, 3600]
```

# plt.plot(x,y,xolor='color name', linestyle='line_style', linewidth=value, marker='marker symbol', label='label name')

```python
plt.plot(months, sales, color='hotpink', linestyle='--', linewidth=2, marker='h', label='Monthly Sales Data')
plt.xlabel=('months')
plt.ylabel=('sales per month')
plt.title=('Monthly Sales Report')
plt.xlim(0,4)
plt.ylim(2000,4000)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(0,2,100)
plt.plot(x,x,label='Linear')
plt.plot(x,x**2,label='Quadratic')
plt.plot(x,x**3,label='Cubic')
plt.xlabel('x lines')
plt.ylabel('y lines')
plt.title('Simple Lines Plot')
plt.legend(loc='upper left')
plt.show()
```

# Bar Plot Example
## plt.bar(x,y,color=", edgecolor=' ')

```python
import matplotlib.pyplot as plt
```

```python
categories=['Electronics', 'Mobiles', 'Books', 'Groceries', 'Home Decoration']
sales=[50000,60000,30000,90000,40000]
bar_colors=['red','blue','green','purple','orange']
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color=bar_colors, edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Sales Amount (MYR)')
plt.title('Categories wise Sales Report')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
categories=['Electronics', 'Mobiles', 'Books', 'Groceries', 'Home Decoration']
sales=[50000,60000,30000,90000,40000]
plt.figure(figsize=(8,5))
bar_colors=['khaki','skyblue','lime','indigo','gold']
plt.bar(categories,sales,color=bar_colors, edgecolor='black', label='Sales Info')
plt.xlabel('Categories')
plt.ylabel('Sales Amount (MYR)')
plt.title('Categories wise Sales Report')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xticks(rotation=30, fontsize=12, color='magenta')
plt.yticks([20000,40000,60000,80000,100000], ['20k','40k','60k','80k','1M'], fontsize=12, color='olive')
plt.legend(title='Legend',fontsize=8)
plt.tight_layout()
plt.show()
```

# Scatter Plot
## plt.scatter(x,y,s,=", c='colorName', marker='markername', alpha=transp., edgecolor='color name', label='label of legend')

```python
height_cm=[150,155,160,165,170,175,180,185]
weight_kg=[50,53,56,59,62,66,68,72]
plt.figure(figsize=(8,5))
plt.scatter(height_cm,weight_kg, s=100, c='indigo', marker='*', alpha=0.9, edgecolor='black', label='weight vs height')
plt.title('Scatter Plot Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.xlim(140,190)
plt.xticks(rotation=0, fontsize=10, color='maroon')
plt.yticks([50,55,60,65,70,75,80,85], ['50kg','55kg','60kg','65kg','70kg','75kg','80kg','85kg'], fontsize=10, color='maroon')
plt.grid(color='lightpink', alpha=0.7, linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
