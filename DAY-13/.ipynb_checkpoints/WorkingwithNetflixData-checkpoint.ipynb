```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('netflix_data.csv')
print(df.head())
```

```python
df=df.dropna(subset=['release_year'])
```

```python
df['release_year']=df['release_year'].astype(int)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['release_year'],bins=100,color='teal', edgecolor='black', alpha=0.8)
plt.title('Netflix released Title by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(color='grey', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('netflix_data.csv')
df=df.dropna(subset=['release_year'])
df['release_year']=df['release_year'].astype(int)
plt.figure(figsize=(8,5))
plt.hist(df['release_year'],bins=100,color='azure', edgecolor='blue', alpha=0.8)
plt.title('Netflix released Title by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(color='black', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.savefig('netflix_bar.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
df=df.dropna(subset=['release_year','type'])
```

```python
df['release_year']=df['release_year'].astype(int)
```

```python
movies=df[df['type']=='Movie']
shows=df[df['type']=='TV Show']
```

```python
plt.figure(figsize=(10,5))
plt.hist([df['release_year'], shows['release_year']],
         bins=30,
         color=['black','pink'], 
         edgecolor='blue', 
         alpha=0.8,
         label=['Movies','TV-Shows']
         )
plt.title('Netflix released Title by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles Released')
plt.grid(color='black', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.savefig('netflix_showvsmovies.jpg', dpi=300, bbox_inches='tight')
plt.show()
```

# Use netflix_data.csv dataset
    ## print how many movies and tv shows are available on Netflix
    ## display movies and tv shows proportion in a pie plot
    ## save file as pdf

```python
df=df.dropna(subset=['release_year'])
```

```python
type_counts=df['type'].value_counts()
print(type_counts)
```

```python
plt.figure(figsize=(8,5))
plt.pie(type_counts,
         labels=type_counts.index,
         colors=['orange','green'],
         autopct='%1.2f%%', startangle=140, shadow=True, explode=(0.05,0.05)
         )         
plt.title('Netflix Movies and TV Shows Distribution')
plt.tight_layout()
plt.legend()
plt.savefig('netflix_pieplot.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
