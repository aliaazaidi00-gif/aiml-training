# Scatter Plots
### In scatter plots we represent data in points, where each point represent a data observation. We can check how to value correlate.

```python
product_sales={'Laptop':750,'Smart Phone':650,'Smart Watch':400,'Tablets':500,'AC':400}
products=list(product_sales.keys())
sales=list(product_sales.values())
plt.scatter(products,sales)
plt.title('Basic Product and Sales Represenation')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
players=['Gakpo','Pedro','Semenyo','Cucurella','Haaland']
match1_score=[45,89,76,90,51]
match2_score=[30,16,83,67,90]
match3_score=[78,90,32,43,71]
plt.figure(figsize=(8,5))
plt.scatter(players,match1_score,color='red',label='Match One',s=100, marker='p')
plt.scatter(players,match2_score,color='yellow',label='Match Two',s=100, marker='x')
plt.scatter(players,match3_score,color='blue',label='Match Three',s=100, marker='h')
plt.title('Comparison of Players in Two Matches')
plt.xlabel('Players')
plt.ylabel('Score')
plt.legend(loc='lower right')
plt.grid(True,linestyle='--',alpha=0.5)
plt.show()
```

## Write a comparison that shows average calories burned for each workout type from final_dataset

```python
df=pd.read_csv("C:/Users/User/Desktop/AIML-TRAINING/DAY-10/Final_data.csv")
avg_calories=df.groupby('Workout_Type')['Calories_Burned'].mean()
selected_index=avg_calories.index
selected_value=avg_calories.values
plt.scatter(selected_index,selected_value,color='red',label='Match One',s=100, marker='p')
plt.title('Comparison of Workout and Calories Burned')
plt.xlabel('Workout Type')
plt.ylabel('Calories Burned')
plt.grid(True,linestyle='--',alpha=0.5)
plt.show()
```
