# Bascis of Matplotlib chart
    ## Pyplot: Interface for creating and managing plot.
    ## Figure: The complete window to create plot.
    ## Axes: The area where data is plotted.
    ## Axis: x line and y line to define scale and ticks.
    ## Plot: the graph of data.

```python
plt.figure(figsize=(8,6))
```

```python
plt.plot(x,y,marker='D',linestyle='-')
```

```python
plt.title('First Line Chart')
```

```python
plt.xlabel('X Label')
```

```python
plt.ylabel('Y Label')
```

```python
plt.grid(True)
```

```python
plt.show()
```

# First example to create a line chart from our data
    ## Student.xlsx for fetching the data.
    ## Generate line chart based on student name and marks.

```python
import pandas as pd 
```

```python
df=pd.read_excel('C:/Users/User/Desktop/AIML-TRAINING/Day-10/students.xlsx')
```

```python
names=df['Name']
marks=df['Marks']
```

```python
plt.plot(names,marks,marker='s',color='green')
plt.title('Student Marks Line Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(names,marks,marker='o',color='orange')
plt.title('Students Progress Chart')
plt.xlabel('--Student Name--')
plt.ylabel('--Student Marks--')
plt.grid(True)
plt.show()
```

```python
print('Solid Line')
plt.figure(figsize=(8,5))
plt.plot(names,marks,marker='o',color='orange',linestyle='-')
plt.title('Students Progress Chart')
plt.xlabel('--Student Name--')
plt.ylabel('--Student Marks--')
plt.grid(True)
plt.show()
```

```python
print('Dashed Line')
plt.plot(names,marks,marker='o',color='orange',linestyle='--')
plt.show()
```

```python
print('Dotted Line')
plt.plot(names,marks,marker='o',color='orange',linestyle=':')
plt.show()
```

```python
print('Dashed Dotted Line')
plt.plot(names,marks,marker='o',color='orange',linestyle='-.')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(names,marks,marker='v',color='blue',linestyle='-.')
plt.title('Students Progress Chart')
plt.xlabel('--Student Name--')
plt.ylabel('--Student Marks--')
plt.grid(True)
plt.show()
```
