```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/User/Desktop/AIML-TRAINING/DAY-13/netflix_data.csv")
df.head(3)
```

```python
sns.set_theme(style='darkgrid', palette='dark')
sns.countplot(data=df, x='type', hue='type')
plt.title('Number of Movies vs Number of TV shows')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

```python
df=df.dropna(subset=['type','genre'])
```

```python
df=df.dropna(subset=['type','genre'])
sns.countplot(data=df.head(20), x='type', hue='country', palette='Set1')
plt.title('Netflix Content Type by Country')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.legend(title='Country')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(data=df.head(10), x='type', hue='country', palette='Set2')
plt.title('Netflix Content Type by Country only top 10')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.legend(title='Country', bbox_to_anchor=(1.02,1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='type', hue='country', palette='Paired')
plt.title('Netflix Content Type by Country only top 10')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.tight_layout
plt.legend(title='Country', bbox_to_anchor=(1.05,1), loc= 'upper left', ncol=5)
plt.show()
```

# Use Netflix dataset to visualize the distribution of release years for Moives vs TV Shows using Seaborn's histplot()
## Use different color for Movie vs TV Show

```python
df.head()
netflix_year = df[['type', 'release_year']].dropna()
plt.figure(figsize=(8,7))
sns.histplot(data=netflix_year, x='release_year', multiple='dodge', hue='type', palette='Set2',bins=20)
plt.title('Distribution of Release Years: Movies vs TV Shows')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```
