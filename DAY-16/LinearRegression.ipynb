## 1. Linear Regression

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
data=pd.read_csv('sales_data.csv')
data.head()
```

```python
x=data[['Advertising_Spend']]
y=data[['Sales_Revenue']]
```

```python
#step4: Train the model
model=LinearRegression()
model.fit(x,y)
```

```python
#step5: Predict
new_spend=int(input('Enter New Spend to Predict Sales Revenue'))
predicted_revenue=model.predict([[new_spend]])
print(f'New Spend {new_spend} Predicted Revenue {predicted_revenue}')
```
