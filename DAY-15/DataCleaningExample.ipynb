```python
import pandas as pd
```

```python
data={'Students':['Ali','Abu','Atan','Ahmad','Ah Hok','Mutu'],
'Marks':[85,None,98,76,None,55]}
```

```python
df=pd.DataFrame(data)
print('Original Data')
print(df)
```

```python
print('Missing Data')
print('Numbers of missing values are:', df.isnull().sum())
```

```python
average_marks=df['Marks'].mean()
print(average_marks)
```

```python
df.fillna(average_marks, inplace=True)
```

```python
print('after fillna() call')
df
```
